```python
collected = dict()
for file in os.listdir(r".\backup"):
    if file.endswith(".json"):
        collected[file.split(".")[0]] = dict()
        with open(r".\backup\\" + file, "r", encoding="utf8") as f:
            data = json.load(f)
            collected[file.split(".")[0]] = data["list"]["Array"]

cleaned = dict()
for lang, handleandwordpairs in collected.items():
    # print(handleandwordpairs)
    for word in handleandwordpairs:
        cleaned.setdefault(word["handle"], dict())[lang] = word["word"]

json.dump(cleaned, open("cleaned.json", "w", encoding="utf8"), ensure_ascii=False, indent=2)
```
