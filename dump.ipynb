```python
import json
import os
import re
import openpyxl
import zipfile
from datetime import datetime
```

```python
xl = openpyxl.load_workbook('Localization-0108.xlsx')
```

```python
translated_dict = dict()
for row in xl.worksheets[0].iter_rows(min_row=1, max_row=xl.worksheets[0].max_row, min_col=1, max_col=5):
    title = row[0].value
    korean = row[4].value
    translated_dict[title] = korean
```

```python
from pprint import pprint as pp
pp(translated_dict)
```

```python
original_data = json.load(open('./backup/CHT.json', encoding='utf-8'))
```

```python
pp(original_data)
```

```python
for elem in original_data["list"]["Array"]:
    if elem["handle"] in translated_dict.keys():
        elem["word"] = translated_dict[elem["handle"]]
```

```python
json.dump(original_data, open('translated_CHT.json', 'w', encoding='utf-8'), ensure_ascii=False, indent=2)
```
