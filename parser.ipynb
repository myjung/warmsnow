```python
import struct
import json
import os
```

```python
file = open("./xbox/KR.dat", mode="rb")
data = file.read()
file.close()
```

```python
data[16*3:]
```

```python
"""data structure
4 bytes - size of link
n bytes by size of link data
3 bytes 00 00 00 - padding
4 bytes - size of data
n bytes by size of data
3 bytes 00 00 00 - padding
"""
pointer = 0
sliced = data[16*3:]
tempcount = 0
while pointer < len(sliced):
    link_size = struct.unpack("<I", sliced[pointer:pointer+4])[0]
    print(link_size)
    link_data = sliced[pointer+4:pointer+4+link_size].decode("utf-8")
    print(link_data)
    pointer = pointer + 4 + link_size + 3
    data_size = struct.unpack("<I", sliced[pointer:pointer+4])[0]
    print(data_size)
    data_data = sliced[pointer+4:pointer+4+data_size]
    print(data_data.decode("utf-8"))
    pointer = pointer + 4 + data_size + 3
    tempcount += 1
    if tempcount > 1:
        break
```
